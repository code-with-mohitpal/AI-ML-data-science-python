```python
import matplotlib.pyplot as plt

marks = [12,15,15,18,20,20,22,25,25,28]

# Define bins explicitly
bins = [10, 16, 21, 26, 30]  # 10-15,16-20,21-25,26-30

plt.hist(marks, bins=bins,color='red', edgecolor='black')
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("/content/insurance.csv")
```

```python
plt.plot(df['smoker'], df['charges'], color='r', marker='*')
plt.title('Insurance Charges w.r.t to BMI')
plt.xlabel('Bmi')
plt.ylabel('Charges')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.yticks(rotation=30)
plt.show()
```

```python
plt.scatter(df['bmi'], df['charges'], color='blue', marker='o')
plt.title('Insurance Charges w.r.t to BMI')
plt.xlabel('Bmi')
plt.ylabel('Charges')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(df['age'], bins=10, color='orange', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
