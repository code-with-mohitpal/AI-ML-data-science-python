# Part A: Introducing **NumPy**

### Why **Python** is used for AI?
* Artificial Intelligence systems rely heavily on numerical computation, linear algebra, and large-scale data processing.
* While Python is an interpreted language and slower than C or C++, it achieves high performance in AI through optimized numerical libraries such as **NumPy**, **SciPy**, and other deep learning frameworks.
* Python acts as a *control language*, while performance-critical operations are executed in compiled backends

## Why Python's Built-in Structures Aren't Enough for Large Data
* Lists, tuples, sets, dictionaries are great for general programming.
* But for numerical operations on large datasets, they can be slow and memory-inefficient.
* **Example**: Summing millions of numbers in a Python list vs. a specialized array in optimized time-space complexity

### Data Structures for AI
* **Scalar**: A single numeric value
* **Vector**: One-dimensional array of features
* **Matrix**: Two-dimensional data representation
* **Tensor**: Multi-dimensional array (images, sequences, batches)


      Scalar ▶ Vector ▶ Matrix ▶ Tensor

## **NumPy (Numerical Python)**:
* Core Object: **ndarray** (N-dimensional array).
* **Homogeneous** (all elements of the same data type).
* Fixed size at creation.
* Highly optimized for numerical operations.
* **Vectorization**: Perform operations on entire arrays without explicit loops (much faster).
* **Why it's faster**: Implemented in C, optimized for memory layout.
* Foundation for *Pandas*.

### **Vectorization**

**Vectorization**
* *eliminates **explicit loops*** and
* *delegates computation to **optimized libraries***

````python
# Loop
a = np.array([1, 2, 3, 4, 5])
s = 0
for i in range(len(a)):
  s += a[i]
print('Loop based sum:', s)

# Vectorized
s = a.sum()
print('Vectorized sum:', s)
````

````python
#UNDERSTANDING PURPOSE ONLY
# observe the time taken to execute both codes

import time
a = np.array([1, 2, 3, 4, 5])

# Loop
start_time = time.perf_counter() # Why perf_counter()? High‑resolution timer. Includes time elapsed during sleep. Suitable for benchmarking small code snippets.
# perf_counter() returns a floating‑point number representing the value of a high‑resolution performance counter in seconds.
s = 0
for i in range(len(a)):
  s += a[i]
end_time = time.perf_counter()
time_taken = end_time - start_time
print('Loop based sum:', s, 'in', time_taken,'seconds')

# Vectorized
start_time = time.perf_counter()
s = a.sum()
end_time = time.perf_counter()
time_taken = end_time - start_time
print('Vectorized sum:', s, 'in', time_taken,'seconds')
````
