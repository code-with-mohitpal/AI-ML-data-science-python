```python
s=pd.Series([1,2,3,4,5,6,7,8,9])
print(s)
print(type(s))
# series(1D labeled array)
# index
print(s[0])
print(s[1])
```

```python
# Series in pandas
s=pd.Series([23,24,25,26,27])
print(s)
print(type(s))
# Indexing
print(s[0])   #23
print(s[2])    #25
print(s.index)  #all labels
```

```python
# custom indexing 
s2=pd.Series([23,24,25,26], index=["Adam","Eve","Charlie","Bob"])
print(s2["Eve"])
print(s2["Bob"])
```

```python
# Vectorized operations
s1=pd.Series([1,2,3])
s2=pd.Series([4,5,6])
print(s1+s2)
```

```python
# Mutable Value but immutable size
s=pd.Series([1,2,3,4,5,6])
print(s)
s[0]=100
print(s)
changed_s=s.drop(1)
print(changed_s)
```

```python
# secect rows (by index)
df.iloc[0]
df.iloc[4:7]
```

```python
# select rows and columns
df.loc[0,"country"]
df.iloc[0,2]
df.iloc[0,1:5]
df.loc[0,["country","aqi"]]
df.loc[0:5,["country","city","latitude"]]
df.iloc[0:5,1:3]
```

```python
# select single scalar value
df.at[0,"city"]
df.iat[0,2]
df
```

```python
# filtering data
df[df["aqi"]>100]
df[df["aqi"]>100][["city","aqi"]]
df[(df["aqi"]>100) & (df["timestamp"]=="2025-11-04 18:25:17.554219")]
```

```python
aqi_data=df[(df["aqi"]>100) & (df["timestamp"]=="2025-11-04 18:25:17.554219")][["city","aqi"]]
print(aqi_data)
```

```python
# difference in loc and iloc both given same value
aqi_data.loc[360]
aqi_data.iloc[1]
```

```python
# Query return a copy,not a view
df.query("aqi > 100 & timestamp == '2025-11-04 18:25:17.554219'")
my_country="IN"
df.query("country == @my_country")
```
