```python
import pandas as pd
import numpy as np
```

```python


# Load Iris dataset
iris = pd.read_csv("IRIS.csv")
print(iris.head(10))
print(iris.tail(10))
```

```python
print("Shape:", iris.shape)
print("\nData Types:\n", iris.dtypes)
```

```python
print(iris.describe())
```

```python
filtered_iris = iris[
    (iris["petal_length"] > 4.5) &
    (iris["species"] == "Iris-virginica")
]

print(filtered_iris)
```

```python
grouped_iris = iris.groupby("species").agg(
    avg_sepal_length=("sepal_length", "mean"),
    max_petal_width=("petal_width", "max"),
    std_sepal_width=("sepal_width", "std")
)

print(grouped_iris)
```

```python
iris["petal_ratio"] = iris["petal_length"] / iris["petal_width"]

avg_petal_ratio = iris.groupby("species")["petal_ratio"].mean()
print(avg_petal_ratio)
```

```python
titanic = pd.read_csv("Titanic-Dataset.csv")
print(titanic)
```

```python
selected_cols = titanic[["Name", "Sex", "Age", "Fare", "Survived"]]
print(selected_cols.head())
```

```python
filtered_passengers = titanic[
    (titanic["Sex"] == "female") &
    (titanic["Fare"] > 30)
]

print(filtered_passengers)
```

```python
grouped_titanic = titanic.groupby("Pclass").agg(
    survival_rate=("Survived", "mean"),
    average_fare=("Fare", "mean"),
    average_age=("Age", "mean")
)

print(grouped_titanic)
```

```python
grouped_titanic["survival_rate"] *= 100
```
